# MINI PROJECT : MOVIE RATING INSIGHTS

```python
# Importing the necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a sample movie rating dataset

data = {
    'Movie' : ['inception','Avengers','Titanic','Interstellar','Joker','Leo','Vikram','GoodBadUgly','Jailer','Vadachennai'],
    'Genre' : ['Sci-Fi','Action','Romantic','Sci-Fi','Drama','Action','Action','Drama','Action','Crime'],
    'Rating' : [8.8,8.0,7.8,8.6,8.4,7.2,8.3,8.8,7.1,8.5],
    'Votes' : [2300000, 1900000, 1800000, 1900000, 1700000, 950000, 1200000, 800000, 1000000, 750000]
}
```

```python
# Convert the movie data dictionary into a Pandas DataFrame

df = pd.DataFrame(data)
```

```python
# Displaying the dataset

print("\nMovie Rating DataSet :\n")
print(df)
print("\n")
```

```python
# Average rating of all movies

average_rating = df['Rating'].mean()
print(f"\nAverage Movie Rating: {average_rating:.2f}\n")
```

```python
# Top 3 highest rated movies

top_movies = df.sort_values(by='Rating', ascending=False).head(3)
print("\nTop 3 Highest Rated Movies:\n")
print(top_movies[['Movie', 'Rating']])
print("\n")
```

```python
# Genre-wise average rating

genre_avg = df.groupby('Genre')['Rating'].mean().reset_index()
print("\nGenre-wise Average Rating:\n")
print(genre_avg)
print("\n")
```

```python
# Visualizations
# Visualization for movie rating

plt.figure(figsize=(10,6))
sns.barplot(x='Movie', y='Rating', data=df, palette='viridis')
plt.xticks(rotation=45)
plt.title('Movie Ratings')
plt.show()
```

```python
# Visualization of average rating by genre

plt.figure(figsize=(8,6))
sns.barplot(x='Genre', y='Rating', data=genre_avg, palette='magma')
plt.title('Average Rating by Genre')
plt.show()
```

```python
# Comparing Votes Vs Rating

plt.figure(figsize=(10,6))
sns.scatterplot(x='Votes', y='Rating', data=df, hue='Genre', s=100)
plt.title('Votes vs Rating')
plt.show()
```

```python
# Most popular movie (by votes)

most_popular = df[df['Votes'] == df['Votes'].max()]
print("\nMost Popular Movie (By Votes):\n")
print(most_popular[['Movie', 'Votes']])
print("\n")
```
